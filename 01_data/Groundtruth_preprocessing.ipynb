```python
import numpy as np
import tifffile

for i, s in enumerate(source):
    # print(source)
    fsn = source_files[i].split('/')[-1]
    
    # write two channel file
    out_file = out_path_cortex_two_channels + fsn
    im = source[i][:,:,1:]
    im = np.transpose(im, (2, 0, 1))
    im = im.astype(np.uint16)

    tifffile.imsave(out_file, im)
    
    
    # write composited file
    # composite by weights 0.5 and 0.5 for the two channels
    out_file = out_path_cortex_composited + fsn
    im = np.sum(source[1][:,:,1:].astype(np.float) * np.ones((1, 1, 2)) / 2., 2)
    im = im.astype(np.uint16)

    tifffile.imsave(out_file, im)

    
    # write the astrocytes raw data
                     
    out_file = out_path_astro + fsn
    im = source[i][:,:,0]
    im = im.astype(np.uint16)

    tifffile.imsave(out_file, im)
    
```
